```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Grafico de linea (relacion entre dos variables continuas y tendencias o secuencias ordenadas)
x = [1,2,3,4,5]
y = [2,3,5,7,9]

plt.plot(x,y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Grafico de linea')

plt.show()
```

```python
# Grafico de dispersion (para patrones y visualizacion de datos en ciencia de datos)
x = [1,2,3,4,5]
y = [2,3,5,7,9]

plt.scatter(x,y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Grafico de dispersion')
plt.show()
```

```python
# Histograma (ver la distribucio nde variable numerica, frecuencia de datos en intervalos, detectar valores atipicos)
import numpy as np

datos = np.random.rand(1000)#retorna aleatoriamente 1000 numeros

plt.hist(datos,bins=20,color='skyblue',edgecolor='black')#bins cantidad de intervalos
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Histograma')

plt.show()
```

```python
# Grafico de barras

categoria = ['A', 'B', 'C']
valor = [150,230,180]

plt.bar(categoria,valor,color='pink',edgecolor='black')
plt.xlabel('Categorias')
plt.ylabel('Valor')
plt.title('Grafico de barras')
plt.show()
```
